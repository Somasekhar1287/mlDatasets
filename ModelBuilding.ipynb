```python
from sklearn.metrics import classification_report
```

```python
y_true=[1,1,1,0,0,0,1,0,0,1]
y_pred=[1,1,0,1,0,0,1,0,0,0]
```

```python
target_names = ['True', 'False']
```

```python
print(classification_report(y_true, y_pred, target_names=target_names))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_true, y_pred)

# Step 3: Plot the confusion matrix
# Use ConfusionMatrixDisplay to create a visually appealing plot
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=target_names)

# Plot and show the figure
disp.plot(cmap=plt.cm.Blues) # Using a blue colormap (optional)
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
y_true = [3, -0.5, 2, 7]
```

```python
y_pred = [2.5, 0.0, 2, 8]
```

```python
'''
=|3-2.5|+|-0.5-0.0|+|2-2|+|7-8|
=|0.5|+|-0.5|+|0|+|1|
=0.5+0.5+1
=2
2/4=0.5
'''
```

```python
mean_absolute_error(y_true, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
```

```python
'''
=(0.5)^2+(0.5)^2+(0)^2+(1)^2
=0.25+0.25+0+1
=0.50+1
=1.5
=1.5/4=0.375
'''
```

```python
mean_squared_error(y_true, y_pred)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
root_mean_squared_error(y_true, y_pred)
```
