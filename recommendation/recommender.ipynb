```python
# product_id : weight
user1 = {1: 5, 3: 2, 5: 1}
user2 = {1: 4, 2: 1, 5: 2}
```

```python
def cosine_similarity_sparse(v1, v2):
    dot = 0.0

    for key in v1:
        if key in v2:
            dot += v1[key] * v2[key]

    norm1 = math.sqrt(sum(val * val for val in v1.values()))
    norm2 = math.sqrt(sum(val * val for val in v2.values()))

    if norm1 == 0 or norm2 == 0:
        return 0.0

    return dot / (norm1 * norm2)
score=cosine_similarity_sparse(user1, user2)
```

```python
def similarity_relation(score):
    if score >= 0.85:
        return "very strong"
    elif score >= 0.65:
        return "strong"
    elif score >= 0.40:
        return "medium"
    elif score >= 0.20:
        return "weak"
    else:
        return "none"
```

```python
similarity_relation(score)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer(stop_words="english")

tfidf_matrix=tfidf.fit_transform(data.Category)

tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)

movie_index = pd.Series(data.index, index =data ['Titles']).drop_duplicates()
```

```python
movie_index
```

```python
def get_rec_movies(movie_name,num):
    movie_id=movie_index[movie_name]
    cosine_scores=list(enumerate(cosine_sim_matrix[movie_id]))
    
    cosine_scores = sorted(cosine_scores, key=lambda x:x[1], reverse = True)
    
    cosine_scores_N = cosine_scores[0: num+1]
    
    movie_idx  =  [i[0] for i in cosine_scores_N]
    movie_scores =  [i[1] for i in cosine_scores_N]
    
    movie_similar_show = pd.DataFrame(columns=["name", "Score"])
    movie_similar_show["name"] = data.loc[movie_idx, "Titles"]
    movie_similar_show["Score"] = movie_scores
    movie_similar_show.reset_index(inplace = True)
    print (movie_similar_show)


```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
Stopwords=stopwords.words('english')
```

```python
import re
def clean_text(text):
    text = re.sub(r"-", "", text)
    text=re.sub(r"[^a-zA-Z0-9]", " ", text)
    words=[]
    for word in text.split(" "):
        if len(word)>3:
            words.append(word)
    w=[word for word in words if word not in Stopwords]
    return (" ".join(w))
```

```python
data.Category=data.Category.apply(clean_text)
```
