```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
```

```python
housing = fetch_california_housing()
```

```python
print(housing.data.shape, housing.target.shape)
```

```python
housing
```

```python
x=pd.DataFrame(housing.data)
```

```python
x.columns=housing.feature_names
```

```python
x
```

```python
y=pd.DataFrame(housing.target,columns=["target"])
```

```python
y
```

```python
housing_data=pd.concat([x,y],axis=1)
```

```python
housing_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(housing_data.corr())
```

```python
housing_data.isna().sum()
```

```python
for col in housing_data.columns:
    if col!="target":
        sns.boxplot(housing_data[col]);plt.title(col);plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc_data=sc.fit_transform(housing_data.iloc[:,:-1])
```

```python
sc_data=pd.DataFrame(sc_data,columns=housing_data.iloc[:,:-1].columns)
```

```python
sc_data
```

```python
data=pd.concat([sc_data,housing_data.iloc[:,-1]],axis=1)
```

```python
data
```

```python
import statsmodels.formula.api as smf
```

```python
model1=smf.ols('target ~MedInc+HouseAge+AveRooms+AveBedrms+Population+AveOccup+Latitude+Longitude',data=data).fit()
model1.summary()
```

```python
model2=smf.ols('target ~MedInc+HouseAge+AveRooms+AveBedrms+Population+AveOccup',data=data).fit()
model2.summary()
```

```python
import statsmodels.api as sm
from scipy import stats
```

```python
model1.summary()
pred1=model1.predict()
res = model1.resid
sm.qqplot(res)
plt.show()
import pylab
stats.probplot(res, dist = "norm", plot = pylab)
plt.show()

sns.residplot(x = pred1, y = data["target"], lowess = True)
plt.xlabel('Fitted')
plt.ylabel('Residual')
plt.title('Fitted vs Residual')
plt.show()
```

```python
model2.summary()
pred=model2.predict()
res = model2.resid
sm.qqplot(res)
plt.show()
import pylab
stats.probplot(res, dist = "norm", plot = pylab)
plt.show()

sns.residplot(x = pred, y = data["target"], lowess = True)
plt.xlabel('Fitted')
plt.ylabel('Residual')
plt.title('Fitted vs Residual')
plt.show()
```

```python

from sklearn.model_selection import train_test_split
train,test = train_test_split(data, test_size = 0.2) # 20% test data

# preparing the model on train data 
model_train = smf.ols('target ~MedInc+HouseAge+AveRooms+AveBedrms+Population+AveOccup+Latitude+Longitude', data = train).fit()

# prediction on test data set 
test_pred = model_train.predict(test)

# test residual values 
```

```python
test_resid = test_pred - test.target
# RMSE value for test data 
test_rmse = np.sqrt(np.mean(test_resid * test_resid))
test_rmse
```

```python


# train_data prediction
train_pred = model_train.predict(train)

# train residual values 
train_resid  = train_pred - train.target
# RMSE value for train data 
train_rmse = np.sqrt(np.mean(train_resid * train_resid))
train_rmse
```

```python
model_train2 = smf.ols('target ~MedInc+HouseAge+AveRooms+AveBedrms+Population+AveOccup', data = train).fit()

# prediction on test data set 
test_pred = model_train2.predict(test)

# test residual values 
```

```python
test_resid = test_pred - test.target
# RMSE value for test data 
test_rmse = np.sqrt(np.mean(test_resid * test_resid))
test_rmse
```

```python
# train_data prediction
train_pred = model_train2.predict(train)

# train residual values 
train_resid  = train_pred - train.target
# RMSE value for train data 
train_rmse = np.sqrt(np.mean(train_resid * train_resid))
train_rmse
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
rf_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=20,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(train.iloc[:,:-1], train.iloc[:,-1])
```

```python
y_train_pred = rf_model.predict(train.iloc[:,:-1])
y_test_pred = rf_model.predict(test.iloc[:,:-1])
```

```python
def print_metrics(y_true, y_pred, name):
    print(f"\n{name} Metrics")
    print("MAE :", mean_absolute_error(y_true, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print("R2  :", r2_score(y_true, y_pred))

print_metrics(train.iloc[:,-1], y_train_pred, "Train")
print_metrics(test.iloc[:,-1], y_test_pred, "Test")
```

```python
residuals = test.iloc[:,-1] - y_test_pred

plt.figure()
plt.scatter(y_test_pred, residuals)
plt.axhline(0)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Predicted")
plt.show()
```

```python
plt.figure()
plt.scatter(test.iloc[:,-1], y_test_pred)
plt.plot([test.iloc[:,-1].min(), test.iloc[:,-1].max()],
         [test.iloc[:,-1].min(), test.iloc[:,-1].max()])
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted")
plt.show()
```
