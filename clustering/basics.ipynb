```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# 1. Sample data
# Each point (x, y) represents an observation
x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]
data = list(zip(x, y)) # Combine x and y coordinates into a list of points

# 2. Perform hierarchical clustering (compute the linkage matrix)
# 'ward' method minimizes the variance of the clusters being merged
# 'euclidean' is a common distance metric
linkage_data = linkage(data, method='ward', metric='euclidean')

# 3. Plot the dendrogram
plt.figure(figsize=(10, 5)) # Optional: adjust figure size
dendrogram(linkage_data)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
