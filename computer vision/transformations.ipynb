# translation

```python
import numpy as np
import cv2 as cv

img = cv.imread('images//birds//176_eagle-1753002_1280.jpg',
                cv.IMREAD_GRAYSCALE)
assert img is not None, 
"file could not be read, check with os.path.exists()"
rows,cols = img.shape

M = np.float32([[1,0,50],[0,1,100]])
dst = cv.warpAffine(img,M,(cols,rows))

cv.imshow('original',img)
cv.imshow('img',dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

# rotation

```python
#, cv.IMREAD_GRAYSCALE
img = cv.imread('images//birds//174_bird-3158784_1280.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,_ = img.shape
 
# cols-1 and rows-1 are the coordinate limits.
M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),45,1)
dst = cv.warpAffine(img,M,(cols,rows))
```

```python
cv.imshow('original',img)
cv.imshow('img',dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

# affine transformation

```python
img = cv.imread('images//birds//174_bird-3158784_1280.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv.getAffineTransform(pts1,pts2)

dst = cv.warpAffine(img,M,(cols,rows))
cv.imshow('img',dst)
cv.waitKey(0)
```

```python

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv.imread('images//birds//163_bird-4051412_1280.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv.getPerspectiveTransform(pts1,pts2)

dst = cv.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
