# face Detection

```python
import cv2
import numpy as np

```

```python
# Load Haar Cascade classifier
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

# Read image
img = cv2.imread("IMG_20180519_074322716_HDR.jpeg")
img=cv2.resize(img,(640,480))

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.3,
    minNeighbors=5,
    minSize=(30, 30)
)

# Draw rectangles
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Show result
cv2.imshow("Face Detection", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 as cv
har_path="haarcascade_frontalface_alt.xml"
face_path="IMG_20180519_074322716_HDR.jpeg"
face_img=cv.imread(face_path)
face_img=cv.resize(face_img,(640,480))
face_cascades=cv.CascadeClassifier(har_path)

def detect_faces(face_img):
    gray_img=cv.cvtColor(face_img,cv.COLOR_BGR2GRAY)
    faces=face_cascades.detectMultiScale(gray_img,scaleFactor=1.6)
    for (x,y,width,height) in faces:
            cv.rectangle(face_img,(x,y),(x+width,y+height),(0,255,0),5)
    face_box=img[y:y+height,x:x+width]
    face_coords=[x,y,width,height]
    return face_img,face_box,face_coords
bound_img,_,_=detect_faces(face_img)
cv.imshow('face bounding box',bound_img)
cv.waitKey(0)
```

```python
import cv2

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=5,
        minSize=(30, 30)
    )

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

    cv2.imshow("Live Face Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
