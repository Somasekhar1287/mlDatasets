# Draw line

```python
import cv2 as cv
import sys

img = cv.imread("images//birds//020_beach-1852945_640.jpg")

if img is None:
    sys.exit("Could not read the image.")

#cv.line(img,(0,50),(100,200),(255,0,0),2)
cv.line(img,(0,50),(250,50),(0,0,255),2)
cv.imshow("Display window", img)
k = cv.waitKey(0)

if k == ord("s"):
    cv.imwrite("hummingbird.jpg", img)
```

# circle

```python
img = cv.imread("images//birds//029_kingfisher-3159334_640.jpg")

if img is None:
    sys.exit("Could not read the image.")

cv.circle(img,(100,100),90,(0,10,200),5)
cv.imshow("Display window", img)
k = cv.waitKey(0)

if k == ord("s"):
    cv.imwrite("hummingbird.jpg", img)
```

# Rectangle

```python
img = cv.imread("images//birds//030_kingfisher-2363879_640.jpg")

if img is None:
    sys.exit("Could not read the image.")

h,w,c=img.shape

cv.rectangle(img,(w//2,h//2),((w//2)-100,(h//2)-100),(255,0,0),2)

cv.imshow("Display window", img)
k = cv.waitKey(0)

if k == ord("s"):
    cv.imwrite("hummingbird.jpg", img)
```

# Text 

```python
img = cv.imread("images//birds//030_kingfisher-2363879_640.jpg")

if img is None:
    sys.exit("Could not read the image.")

h,w,c=img.shape

cv.rectangle(img,(w//2,h//2),((w//2)-100,(h//2)-100),(255,0,0),2)
#cv.putText(img,"King Fisher",((w//2)-120,(h//2)-120),cv.FONT_HERSHEY_COMPLEX,2,(0,20,250),2)
cv.putText(img,"King Fisher",((w//2)-120,(h//2)-120)
           ,cv.FONT_HERSHEY_COMPLEX,1,(0,20,250),1,cv.LINE_AA)

cv.imshow("Display window", img)
k = cv.waitKey(0)

if k == ord("s"):
    cv.imwrite("hummingbird.jpg", img)
```
