```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
email=pd.read_csv("datasets/navie bayes/email.csv")
```

```python
email=email.iloc[:-1,:]
```

```python
email["Category"].value_counts()
```

```python
email_ham = email[email['Category'] == 'ham']
```

```python
email_ham
```

```python
email_spam = email[email['Category'] == 'spam']
```

```python
email_spam
```

```python
email_ham = email_ham.sample(frac=1)
```

```python
email_ham
```

```python
email_ham.iloc[:747,:]
```

```python
email_data=pd.concat([email_ham.iloc[:747,:],email_spam],axis=0)
```

```python
email_data
```

```python
email_data.isna().sum()
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
Stopwords=stopwords.words('english')
```

```python
import re
def clean_text(text):
  text=re.sub("[^a-zA-Z" "]+"," ",text).lower()
  text=re.sub("[0-9" "]+"," ",text)
  words=[]
  for word in text.split(" "):
    if len(word)>3:
      words.append(word)
  w=[word for word in words if word not in Stopwords]
  return (" ".join(w))
```

```python
email_data["Message"]=email_data["Message"].apply(clean_text)
```

```python
email_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(
    ngram_range=(1, 2), # unigrams + bigrams
    max_features=5000)
tfidf_matrix = vectorizer.fit_transform(email_data["Message"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(email_data.iloc[:,-1],email_data.iloc[:,0],test_size=0.2)
```

```python
y_train
```

```python
tfidf_train_matrix = vectorizer.fit_transform(x_train)
```

```python
tfidf_test_matrix = vectorizer.transform(x_test)
```

```python
tfidf_train_matrix.shape
```

```python
tfidf_train_matrix.toarray()
```

```python
tfidf_test_matrix.shape
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
```

```python
classifier_mb = MB()
```

```python
classifier_mb.fit(tfidf_train_matrix, y_train)
```

```python
# Evaluation on Test Data
test_pred_m = classifier_mb.predict(tfidf_test_matrix)
accuracy_test_m = np.mean(test_pred_m == y_test)
print("Accuracy on test: ",accuracy_test_m)
################ train data
```

```python
train_pred_m=classifier_mb.predict(tfidf_train_matrix)
accuracy_train_m=np.mean(train_pred_m==y_train)
print("Train accuracy: ",accuracy_train_m)
```

```python

print("Test cross table:  ",pd.crosstab(test_pred_m,y_test))
```

```python

print("Train cross table: ",pd.crosstab(train_pred_m,y_train))
```
